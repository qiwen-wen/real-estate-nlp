```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Read Data

```python
df = pd.read_csv("../data/processed/listing_sample.csv")
df
```

## Analyzing Remark Pattern

```python
df['word_count'] = df['remarks'].str.len()
df.head()
```

```python
df['word_count'].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Remarks Word Count")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()
```

## Conclusion of Remark Pattern
As shown in the graph, the remark pattern is a skewed to the right with few outliers
